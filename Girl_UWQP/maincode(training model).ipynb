```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('water_quality_main.csv')
dataset
```

```python
dataset.isnull().any()
```

```python
dataset.describe()
```

```python
dataset=dataset.drop('Unnamed: 0',axis=1)
dataset
```

```python
x=dataset.iloc[:,0:7].values
x
```

```python
dataset['station'].value_counts()
```

```python
y=dataset.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(criterion='mse',n_estimators=20,random_state=0)
```

```python
rf.fit(x_train,y_train)
```

```python
from joblib import dump
dump(rf,'random.save')
```

```python
y_predict=rf.predict(x_test)
y_predict
```

```python
y_test
```

```python
rf.predict([[1393.0,6.7,7.50,203.0,1.8965,0.100,27.0]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```

```python
from sklearn.tree import DecisionTreeRegressor
dataa=DecisionTreeRegressor(criterion='mse',random_state=0)
dataa.fit(x_train,y_train)
```

```python
y_pred1=dataa.predict(x_test)
y_pred1
```

```python
dataa.predict([[1393.0,6.7,7.50,203.0,1.8965,0.100,27.0]])
```

```python
r2_score(y_test,y_pred1)
```
